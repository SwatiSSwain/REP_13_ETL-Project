```python
import pandas as pd
import datetime
import numpy as np
```

```python
PI_2015 = ("source_files/Police_Incidents_2015.csv")
PI_2016 = ("source_files/Police_Incidents_2016.csv")
PI_2017 = ("source_files/Police_Incidents_2017.csv")
PI_2018 = ("source_files/Police_Incidents_2018.csv")
PI_2018_2019 = ("source_files/Police_Incidents_Last_2Years.csv")
Neigh_Key = ("target_files/MLPS_Neighborhoods.csv")
```

```python
PI_2015_df = pd.read_csv(PI_2015)
PI_2016_df = pd.read_csv(PI_2016)
PI_2017_df = pd.read_csv(PI_2017)
PI_2018_df = pd.read_csv(PI_2018)
PI_Last2Years_df = pd.read_csv(PI_2018_2019)
Neigh_Key_df = pd.read_csv(Neigh_Key)
```

```python
PI_pre18 = [PI_2015_df,PI_2016_df,PI_2017_df,PI_2018_df]
PI_pre18_merge = pd.concat(PI_pre18)
PI_pre18_merge.head(1)
```

```python
PI_pre18_final = PI_pre18_merge.drop(columns={"PublicAddress", "ControlNbr",
                                                "Precinct", "BeginDate", "Time",
                                                 "EnteredDate", "GBSID", "Lat", 
                                                  "Long", "X", "Y",
                                                 "LastChanged", "LastUpdateDate",
                                                 "Description","UCRCode","FID"})
PI_pre18_final.head(5)
```

```python
#PI_Last2Years_df
```

```python
PI_Last2Years_final = PI_Last2Years_df.drop(columns={"publicaddress","reportedTime", "reportedDateTime",
                                                "precinct", "beginDate", "beginTime",
                                                 "enteredDate", "X", "Y",
                                                 "lastchanged", "LastUpdateDateETL","enteredDate",
                                                 "centerX","centerY","centerLong","centergbsid",
                                                 "centerLat", "enteredDate", "description",
                                                 "OBJECTID", "UCRCode", "lastchanged","LastUpdateDateETL"})

```

```python
PI_pre18_final = PI_pre18_final.rename(columns={"CCN":"caseNumber", "ReportedDate":"reportedDate","Offense":"offense","Neighborhood":"neighborhood"})
```

```python
yearjoins = [PI_pre18_final,PI_Last2Years_final]
PI_Final = pd.concat(yearjoins)
PI_Final.head(5)
```

```python
PI_Final['caseNumber'] = PI_Final['caseNumber'].str.replace(" ","")
PI_Final.head(5)
```

```python
incident_id = PI_Final.index + 1
```

```python
PI_Final['incident_id']= incident_id
```

```python
PI_Final.head(5)
```

```python
PI_Final['neighborhood']= PI_Final['neighborhood'].str.upper()
```

```python
PI_Final = PI_Final.replace(to_replace = "HUMBOLDT INDUSTRIAL AREA", value ="HUMBOLDT INDUST AREA")
PI_Final = PI_Final.replace(to_replace = "CAMDEN INDUSTRIAL", value ="CAMDEN INDUSTRIAL AREA")        
PI_Final = PI_Final.replace(to_replace = "CARAG", value ="South Uptown")
PI_Final = PI_Final.replace(to_replace = "BRYN - MAWR", value ="Bryn Mawr")
PI_Final = PI_Final.replace(to_replace = "COLUMBIA PARK", value ="Columbia")
PI_Final = PI_Final.replace(to_replace = "MID - CITY INDUSTRIAL", value ="Mid City Industrial Ar")
PI_Final = PI_Final.replace(to_replace = "NICOLLET ISLAND - EAST BANK", value ="Nicollet Is/East Bank")
PI_Final = PI_Final.replace(to_replace = "UNIVERSITY OF MINNESOTA", value ="University")
PI_Final = PI_Final.replace(to_replace = "STEVENS SQUARE - LORING HEIGHTS", value ="Steven's Square - Loring Heights")
PI_Final['neighborhood']= PI_Final['neighborhood'].str.upper()

PI_Final.head(5)
```

```python
Neigh_Key_df['name']=Neigh_Key_df['name'].str.upper()
Neigh_Key_df = Neigh_Key_df.rename(columns={'name':"neighborhood"})
Neigh_Key_df
```

```python
PI_Final_ID = pd.merge(PI_Final, Neigh_Key_df, on='neighborhood', how= 'left')
PI_Final_ID.head(5)
```

```python
PI_Final_ID['neighborhood_id'] = PI_Final_ID['neighborhood_id'].values.astype(np.int64)
PI_Final_ID['community_id'] = PI_Final_ID['community_id'].values.astype(np.int64)
PI_Final_ID.dtypes
```
